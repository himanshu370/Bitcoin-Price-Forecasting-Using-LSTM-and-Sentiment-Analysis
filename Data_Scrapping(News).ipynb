```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Load the dataset into a DataFrame
data = pd.read_csv("Crypto News.csv")
```

```python
data.head()
```

```python
# Initialize an empty list to store scraped data
scraped_data = []

# Iterate through each row in the dataset
for index, row in data.iterrows():
    # Get the URL from the current row
    url = row['url']

    # Check if the URL is not missing or invalid
    if pd.notnull(url) and isinstance(url, str) and url.startswith("http"):
        # Send a GET request to the URL
        response = requests.get(url)

        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Get the HTML content of the webpage
            html_content = response.text

            # Parse the HTML content using BeautifulSoup
            soup = BeautifulSoup(html_content, 'html.parser')

            # Extract the title
            title_element = soup.find('title')
            if title_element:
                title = title_element.text.strip()
            else:
                title = "Title element not found."

            # Extract the body content from the meta tag
            meta_description_tag = soup.find('meta', attrs={'name': 'description'})
            if meta_description_tag and 'content' in meta_description_tag.attrs:
                body = meta_description_tag['content'].strip()
            else:
                body = "Body content not found within the meta tag."

            # Append the scraped data to the list
            scraped_data.append({'newsDatetime': row['newsDatetime'], 'title': title, 'body': body})
        else:
            print("Failed to fetch the webpage. Status code:", response.status_code)
            print("Invalid URL:", url)
    else:
        print("Invalid URL:", url)
```

```python
# Convert the list of scraped data into a DataFrame
scraped_df = pd.DataFrame(scraped_data)

# Save the scraped data to a new CSV file
scraped_df.to_csv("scraped_data.csv", index=False)
```
